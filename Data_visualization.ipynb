```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify_songs.csv')
df
```

```python
df['track_album_release_date'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df[['loudness']], bins=10, edgecolor='black', color='skyblue')
plt.title('Histograma')
plt.xlabel('loudness')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['loudness'],df['acousticness'], c='blue', alpha=0.5)
plt.title('Loudness vs Acousticness')
plt.xlabel('loudness')
plt.ylabel('acousticness')
plt.grid(alpha=0.5)
plt.show()
```

```python
genre_dance = df.groupby('playlist_genre')['danceability'].sum()
genre_dance = pd.DataFrame(genre_dance)
```

```python
#Arreglar fechas con el formato incorrecto
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce', format='mixed')

#NaN
df = df.dropna(subset=['track_album_release_date'])

# Sort por Fecha de lanzamiento
df_sorted = df.sort_values('track_album_release_date')

plt.figure(figsize=(8, 5))
plt.plot(df_sorted['track_album_release_date'], df_sorted['loudness'], linestyle='-', marker='o', markersize=4, color='green')
plt.title('Line Plot of Tempo Over Album Release Dates')
plt.xlabel('Album Release Date')
plt.ylabel('Loudness')
plt.grid(alpha=0.5)
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
# Box Plot: Por generos de playlist
df.boxplot(column='danceability', by='playlist_genre', grid=False, patch_artist=True)

plt.title('Box Plot of Danceability by Playlist Genre')
plt.suptitle('')  
plt.xlabel('Playlist Genre')
plt.ylabel('Danceability')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Pie Chart
category_counts = df['playlist_genre'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'pink', 'lightgreen'])
plt.title('Pie Chart')
plt.show()
```
