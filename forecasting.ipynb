```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify_songs.csv')
df
```

```python
df.columns
```

```python
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'])
df['year'] = df['track_album_release_date'].dt.year  # Extract year
```

```python
time_series = df.groupby('year')['track_popularity'].mean().reset_index()
```

```python
X = time_series[['year']]  # Feature
y = time_series['track_popularity'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_series['year'], time_series['track_popularity'], label='Actual Data')
plt.plot(X_train, y_train_pred, label='Training Prediction', linestyle='--')
plt.plot(X_test, y_test_pred, label='Testing Prediction', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Track Popularity')
plt.title('Track Popularity Over Time')
plt.legend()
plt.show()
```
