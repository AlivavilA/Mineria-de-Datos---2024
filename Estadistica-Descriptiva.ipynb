```python
import pandas as pd
```

```python
df = pd.read_csv('aoty.csv')
df
```

```python
df['rating_number'] = df['rating_count'].str.extract(r'(\d[\d,]*)')[0].str.replace(',', '').astype(float)
```

```python
df.describe()
```

```python
df.columns
```

```python
title = df['title'].unique()
artist = df['artist'].unique()
genres = df['genres'].unique()
link = df['album_link'].unique()
```

```python
df.dtypes
```

```python
# Group by 'Entity' and calculate mean
user_score_por_artista = df.groupby('artist')['user_score'].mean()
user_score_por_artista = pd.DataFrame(user_score_por_artista)

# Calculate other statistics like sum, median, etc.
number_ratings_genre = df.groupby('genres')['rating_number'].sum()
number_ratings_genre = pd.DataFrame(number_ratings_genre)
```

```python
user_score_por_artista.sort_values('user_score', ascending=False)
```
